```python
#set up environment for data cleaning from Lobstr

import csv
import pandas as pd
```

```python
#set file path
category1 = 'Chirurgie'
analysed_file_1 = '/Users/etiennecampredon/Downloads/Chir1.csv'
analysed_file_2 = '/Users/etiennecampredon/Downloads/Chir2.csv'
analysed_file_3 = '/Users/etiennecampredon/Downloads/Chir3.csv'
analysed_file_4 = '/Users/etiennecampredon/Downloads/Chir4.csv'
analysed_file_5 = '/Users/etiennecampredon/Downloads/Chir5.csv'
analysed_file_6 = '/Users/etiennecampredon/Downloads/Chir6.csv'
analysed_file_7 = '/Users/etiennecampredon/Downloads/Chir7.csv'
analysed_file_8 = '/Users/etiennecampredon/Downloads/Chir8.csv'
analysed_file_concat = pd.concat(map(pd.read_csv, [analysed_file_1, analysed_file_2, analysed_file_3, analysed_file_4, analysed_file_5, analysed_file_6, analysed_file_7, analysed_file_8]), ignore_index=False)
analysed_file_URL = analysed_file_concat.to_csv('/Users/etiennecampredon/Downloads/Chir.csv')
analysed_file = '/Users/etiennecampredon/Downloads/Chir.csv'
```

```python
#read file with appropriate columns
analysed_data = pd.read_csv(analysed_file, index_col = 'id', usecols = ['id', 'email', 'name', 'address', 'description', 'lat', 'lng', 'category', 'url', 'website'] )
```

```python
#print data head for check
analysed_data.head()
```

```python
#print columns info for check 
analysed_data.columns
```

```python
#check if null data per column 
analysed_data.isnull().sum()
```

```python
# remove duplicates on 3 columns - keep last value from the list

analysed_data = analysed_data.drop_duplicates(keep = 'last', subset = ['name', 'email'])
analysed_data.isnull().sum()
```

```python
#remove null on emails

analysed_data = analysed_data.dropna(axis=0, how='all', subset=['email'])
analysed_data.isnull().sum()
```

```python
#remove all other item than 'salon de coiffure' (or other) from 'category' 

analysed_data = analysed_data.loc[analysed_data['category'].str.contains(category1, case= False, na=False)]
analysed_data.isnull().sum()
```

```python
#add 'new email' column based on 1st email of every 'email' value with split function

analysed_data['new_email'] = analysed_data['email'].str.split(',').str[0]
```

```python
#drop email column to leave new_email only

analysed_data = analysed_data.drop('email', axis=1)
```

```python
#drop rows that contain str in column new emails

analysed_data = analysed_data[analysed_data['new_email'].str.contains('@example|@mail|example@|email@|@email|john@doe|@contact|johndoe|@exemple|exemple@|prenom')==False]
```

```python
#drop duplicates from new_emails (after dropping the rest of emails after the comma)

analysed_data = analysed_data.drop_duplicates(keep = 'last', subset = ['new_email'])
```

```python
analysed_data['new_email'] = analysed_data['new_email'].str.replace('.com.', '.com').replace('.fr.', '.fr').replace('.net.', '.net')
```
